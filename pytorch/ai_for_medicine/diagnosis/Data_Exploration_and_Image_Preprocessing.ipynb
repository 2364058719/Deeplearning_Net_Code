```python
import torch
from torchvision import transforms
from torch.utils.data import DataLoader, Dataset
import cv2

# 自定义 Dataset
class CustomDataset(Dataset):
    def __init__(self, dataframe, directory, x_col, y_col, transform=None):
        self.dataframe = dataframe
        self.directory = directory
        self.x_col = x_col
        self.y_col = y_col
        self.transform = transform

    def __len__(self):
        return len(self.dataframe)

    def __getitem__(self, idx):
        img_name = self.directory + self.dataframe.iloc[idx][self.x_col]
        image = cv2.imread(img_name, cv2.IMREAD_GRAYSCALE)  # 读取灰度图像
        label = self.dataframe.iloc[idx][self.y_col].values.astype(np.float32)

        if self.transform:
            image = self.transform(image)

        return image, label
```

```python
# 图像变换
transform = transforms.Compose([
    transforms.ToPILImage(),
    transforms.Resize((320, 320)),
    transforms.ToTensor(),
    transforms.Normalize(mean=[0.0], std=[1.0])  # 假设初始图像像素范围为 [0, 1]
])
```

```python
# 数据加载
dataset = CustomDataset(
    dataframe=train_df,
    directory="data/nih/images-small/",
    x_col="Image",
    y_col=["Mass"],
    transform=transform
)
```

```python
dataloader = DataLoader(dataset, batch_size=1, shuffle=False)
```

```python
# 取一个样本
generated_image, label = next(iter(dataloader))
print(generated_image.shape)
generated_image = generated_image[0]  # Batch 中取出第一张图像
print(generated_image.shape)
generated_image = generated_image.numpy().squeeze(0)  # 转为 numpy 并移除 channel 维度
print(generated_image.shape)
```

```python
# 绘制预处理后的图像
sns.set_style("white")
plt.imshow(generated_image, cmap='gray')
plt.colorbar()
plt.title('Raw Chest X Ray Image')
print(f"The dimensions of the image are {generated_image.shape[1]} pixels width and {generated_image.shape[0]} pixels height")
print(f"The maximum pixel value is {generated_image.max():.4f} and the minimum is {generated_image.min():.4f}")
print(f"The mean value of the pixels is {generated_image.mean():.4f} and the standard deviation is {generated_image.std():.4f}")
```
